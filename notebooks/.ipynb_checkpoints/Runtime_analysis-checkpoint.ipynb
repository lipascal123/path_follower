## Runtime Analysis

In this script the runtime of different functions are plotted

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# absolute paths to csv files
path = ["/home/pascal/Documents/cc_time.csv",
        "/home/pascal/Documents/predict_path.csv",
        "/home/pascal/Documents/occ_cb_time.csv"]
# import files
predict_path_time = pd.read_csv(path[0]);
cc_time = pd.read_csv(path[1]);
occ_cb_time =pd.read_csv(path[2]);
```

 ### predictPath()

```python
# plot predictPath() runtime
x_pp = list(range(0,predict_path_time.shape[0]))
y_pp = predict_path_time.iloc[:,0]
plt.plot(x_pp,y_pp)
plt.title("predictPath() runtime in seconds")
plt.xlabel("number of datapoints")
plt.ylabel("time in seconds")
```

### collision checker 

```python
# plot collision checker time
x_cc = list(range(0,cc_time.shape[0]))
y_cc = cc_time.iloc[:,0]
plt.plot(x_cc,y_cc)
plt.title("collision checker runtime in seconds")
plt.xlabel("number of datapoints")
plt.ylabel("time in seconds")
```

### occupancy callback timer

```python
# plot occupancy callback time
x_cccb = list(range(0,occ_cb_time.shape[0]))
y_cccb = occ_cb_time.iloc[:,0]
plt.plot(x_cccb,y_cccb)
plt.title("occupancy callback runtime in seconds")
plt.xlabel("number of datapoints")
plt.ylabel("time in seconds")
```

```python
# plot occupancy callback time
x_cccb = list(range(0,cc_time.shape[0]))
y_cccb = cc_time.iloc[:,0] + predict_path_time.iloc[:,0]
plt.plot(x_cccb,y_cccb)
plt.title("occupancy callback runtime in seconds")
plt.xlabel("number of datapoints")
plt.ylabel("time in seconds")
```

```python
# plot occupancy callback time
x_cccb = list(range(0,cc_time.shape[0]))
y_cccb = cc_time.iloc[:,0] + predict_path_time.iloc[:,0] - occ_cb_time.iloc[:,0]
plt.plot(x_cccb,y_cccb)
plt.title("occupancy callback runtime in seconds")
plt.xlabel("number of datapoints")
plt.ylabel("time in seconds")
```
